## 1.1 - Visualization of a Vector  𝑣∈ℝ2
 
You already have seen in the videos and labs, that vectors can be visualized as arrows, and it is easy to do it for a  𝑣∈ℝ2
 , e.g.  𝑣=[13]𝑇
 

The following code will show the visualization.

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_vector(list_v, list_label, list_color):
    _, ax = plt.subplots(figsize=(10 , 10))
    ax.tick_params(axis='x', labelsize=15)
    ax.tick_params(axis='y', labelsize=15)

    ax.set_xticks(np.arange(-10, 10))
    ax.set_yticks(np.arange(-10, 10))

    plt.axis([-10, 10, -10, 10])

    for i , v in enumerate(list_v):
       # .flatten() turns [[1], [3]] into [1, 3], making it easy to grab scalars
       v_flat = v.flatten()
       x_val = float(v_flat[0])
       y_val = float(v_flat[1])
        
       sgn_x = 0.4 * np.sign(x_val)
       sgn_y = 0.4 * np.sign(y_val)
       
       plt.quiver(0, 0, x_val, y_val, color=list_color[i], angles='xy', scale_units='xy', scale=1, label=list_label[i])
       ax.text(x_val - 0.2 + sgn_x, y_val - 0.2 + sgn_y, list_label[i], fontsize=15, color=list_color[i])

    plt.grid()
    plt.gca().set_aspect('equal')
    plt.show()

v = np.array([[1], [3]])

# Arguments: list of vectors as numpy arrays, labels, color 
plot_vector([v], ['v'], ['blue'])
```
