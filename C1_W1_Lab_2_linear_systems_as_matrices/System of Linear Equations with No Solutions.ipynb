## 3 - System of Linear Equations with No Solutions

Given another system of linear equations:

$$
\begin{cases} 
-x_1+3x_2=7, \\ 3x_1-9x_2=1, \end{cases}\tag{2}
$$

Let's find the determinant of the corresponding matrix.

```python
import numpy as np
```

```python
a_2 = np.array([
    [-1, 3],
    [3, -9]
])
print(a_2)
```

```python
a_3 = np.array([7,1])
print(a_3)
```

```python
#reshape this a_3 vector

a_3 = a_3.reshape(2, 1)
print(a_3)
```

```python
#find the determination of two matrices
det_a_2 = np.linalg.det(a_2)

#this system has no solution because the determinant is zero
print(det_a_2)
```

```python
# now i can solve the system of equations using np.linalg.solve
try:
    solution = np.linalg.solve(a_2, a_3)
except np.linalg.LinAlgError as e:
    print("Error:", e)
```
