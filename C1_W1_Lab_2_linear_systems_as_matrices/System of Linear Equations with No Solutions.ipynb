## 3 - System of Linear Equations with No Solutions

Given another system of linear equations:

$$
\begin{cases} 
-x_1+3x_2=7, \\ 3x_1-9x_2=1, \end{cases}\tag{2}
$$

Let's find the determinant of the corresponding matrix.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_2 = np.array([
    [-1, 3],
    [3, -9]
])
print(a_2)
```

```python
a_3 = np.array([7,1])
print(a_3)
```

```python
#reshape this a_3 vector

a_3 = a_3.reshape(2, 1)
print(a_3)
```

```python
#find the determination of two matrices
det_a_2 = np.linalg.det(a_2)

#this system has no solution because the determinant is zero
print(det_a_2)
```

```python
# now i can solve the system of equations using np.linalg.solve
try:
    solution = np.linalg.solve(a_2, a_3)
except np.linalg.LinAlgError as e:
    print("Error:", e)
```

```python
#plot the system of equations   

plt.figure(figsize=(8, 6))
x = np.linspace(-10, 10, 400)   
y1 = (1/3)*x + 7/3
y2 = (1/3)*x + 1/3
plt.plot(x, y1, label='Equation 1: -x + 3y = 7')
plt.plot(x, y2, label='Equation 2: 3x - 9y = 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('System of Linear Equations with No Solutions')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```
