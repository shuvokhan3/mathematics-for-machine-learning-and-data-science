```python
#import NumPy package to access it's functions and objects
import numpy as np 
#import matplotlib package to create the plots
import matplotlib.pyplot as plt
```

System of Linear Equations as Matrices

```python
A = np.array([
    [-1, 3],
    [3, 2]
], dtype=np.dtype(float))

b = np.array([7, 2],dtype=np.dtype(float))

print("Matrix A:")
print(A)
print("Vector b:")
print(b)
```

```python
print(A.shape)
print(b.shape)
```

```python
x = np.linalg.solve(A, b)
print(x)
```

📌 What is the Determinant of a Matrix?

The determinant is a special number that you can calculate from a square matrix (same number of rows and columns).
Why is Determinant Important?
1. ✅ Check if a system has a unique solution

If you're solving equations like:

2x + 3y = 13  
x - y = 1
If determinant ≠ 0 → ✅ Unique solution exists
If determinant = 0 → ❌ No solution or infinite solutions
2. 🔁 Inverse of a matrix
If determinant = 0 → ❌ No inverse
If determinant ≠ 0 → ✅ Inverse exists

👉 This is critical in machine learning, graphics, and physics.

```python
det = np.linalg.det(A)
print("Determinant of A:", det)
```
