## 4 - System of Linear Equations with an Infinite Number of Solutions

By changing the free coefficients of the system $(2)$, you can bring it to consistency:

$$
\begin{cases} 
-x_1+3x_2=7, \\ 3x_1-9x_2=-21, \end{cases}\tag{3}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eq = np.array([
    [-1, 3],
    [3, -9]
])
print(eq)
val = np.array([7, 21])
print(val)
```

```python
#find the determinant of eq
det_eq = np.linalg.det(eq)
print(det_eq)
```

```python
try:
    solve = np.linalg.solve(eq, val)
    print(solve)
except np.linalg.LinAlgError:
    print("The system has no unique solution.")
```

```python
#drow plt of system of equations
x = np.linspace(-10, 10, 100)
y1 = (7 + x) / 3
y2 = (21 - 3 * x) / -9
plt.plot(x, y1, label='Equation 1')
plt.plot(x, y2, label='Equation 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('System of Linear Equations')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```
